```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/s4sauravv/Files/main/Advertising.csv")
```

```python
data.head()
```

```python
fig,axs=plt.subplots(1,3)
data.plot(kind='scatter',x='TV',y='sales',ax=axs[0],figsize=(10,6))
data.plot(kind='scatter',x='radio',y='sales',ax=axs[1])
data.plot(kind='scatter',x='newspaper',y='sales',ax=axs[2])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
x=data[['TV']]
y=data.sales
```

```python
lm.fit(x,y)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
lm.predict([[50]])
```

```python
from sklearn.metrics import r2_score
```

```python
x.head()
```

```python
pridicted_sales =lm.predict(x)
pridicted_sales
```

```python
r2_score(y_true=y,y_pred = pridicted_sales )
```
